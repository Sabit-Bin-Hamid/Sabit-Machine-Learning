# Boxplot

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars['cylinders'].value_counts()
```

```python
cars = cars[cars['cylinders'].isin([4,6,8])]
```

```python
print(cars.shape)
cars['cylinders'].value_counts()
```

```python
sns.boxplot(x = cars['mpg']);
```

```python
#Summary statistics 
cars['mpg'].describe()
```

```python
cars.head()
```

```python
sns.boxplot(x= 'origin', y = 'mpg', data = cars)
```

```python
# implementation of the Tukey's algorithm
def findOutliers(dataset):
    outliers = []
    Q2 = dataset.median()
    Q1 = dataset.describe()[4]
    Q3 = dataset.describe()[6]
    IQR = Q3 - Q1
    lf = Q1 - 1.5*IQR # lf = lower fence
    uf = Q3 + 1.5*IQR # Uf = Upper fence
    for x in dataset:
        if x < lf or x > uf:
            outliers.append(x)
    return outliers
    
```

```python
findOutliers(cars['mpg'])
```

```python
sns.boxplot(x = 'horsepower', data = cars)
```

```python
findOutliers(cars['horsepower'])
```

```python
cars.head()
```

```python
sns.boxplot(x = 'origin', y = 'mpg', hue = 'cylinders', data = cars);
```

```python
sns.boxplot(y = 'mpg', x = 'origin', data = cars)
```

```python
sns.boxplot(x = 'mpg', y = 'origin', data = cars);
```

```python
sns.boxplot(x = 'mpg', y = 'origin', data = cars, order = ['japan', 'europe', 'usa']);
```
